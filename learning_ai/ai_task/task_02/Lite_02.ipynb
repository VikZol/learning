# Задание № 2

**Сравнение чисел с помощью операторов сравнения**

Оператор сравнения "меньше"

```python
# Четыре меньше семи
print(4 < 7)
```

Оператор сравнения "больше"

```python
# Семь больше четырех
print(7 > 4)
```

Оператор сравнения "меньше или равно"

```python
# Четыре меньше или равно семи
print(4 <= 7)
```

Оператор сравнения "больше или равно"

```python
# Семь больше или равно четырем
print(7 >= 4)
```

Оператор сравнения "не равно"

```python
# Четыре не равно семи
print(4 != 7)
```
