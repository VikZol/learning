Подключение API-ключа Tinkoff VoiceKit

```python
import getpass
API_KEY = getpass.getpass('Tinkoff API Key:')
```

Подключение Секретного ключа Tinkoff VoiceKit

```python
SECRET_KEY = getpass.getpass('Tinkof SECRET Key:')
```

```python
from tinkoff_voicekit_client import ClientSTT
from pprint import pprint
client = ClientSTT(API_KEY, SECRET_KEY)
audio_config = {
    "encoding": "MPEG_AUDIO",
    "sample_rate_hertz": 16000,
    "num_channels": 2,
    }
response = client.recognize("/content/drive/MyDrive/Домашние задания AI/Pro/call.mp3",
                            audio_config)
pprint(response)
['results'][0]
```

```python
client = ClientSTT(API_KEY, SECRET_KEY)
audio_config = {
    "encoding": "MPEG_AUDIO",
    "sample_rate_hertz": 16000,
    "num_channels": 2,
}
stream_config = {"config": audio_config}
with open("/content/drive/MyDrive/Домашние задания AI/Pro/call.mp3", "rb") as source:
    responses = client.streaming_recognize(source, stream_config)
    for response in responses:
        pprint(response)
```

```python
client = ClientSTT(API_KEY, SECRET_KEY)
audio_config = {
    "encoding": "MPEG_AUDIO",
    "sample_rate_hertz": 16000,
    "num_channels": 2,
    "max_alternatives": 3
}
stream_config = {"config": audio_config}
with open("/content/drive/MyDrive/Домашние задания AI/Pro/call.mp3", "rb") as source:
    responses = client.streaming_recognize(source, stream_config)
    for response in responses:
        pprint(response)
```
